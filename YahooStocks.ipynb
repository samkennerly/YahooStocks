```python
''' Demonstration of the YahooStocks module '''
import YahooStocks as ys
%matplotlib inline
```

```python
# Use load() to load and align data for several symbols.
# Input a list of symbols Yahoo can recognize.
# Returns a dictionary of DataFrames.

data = ys.load(['SPY','AAPL','^VIX'],start_date='2010-1-1')

print "Each of these is a DataFrame with a DatetimeIndex:"
print data.keys()

data['Adj Close'].plot()

data['Adj Close'].head()
```

```python
# Use get_table() to load all data for one symbol as one DataFrame
IWM = ys.get_table('IWM',start_date='2010-1-1')
IWM.head()
```

```python
# Use make_url() to make a Yahoo! URL without loading anything
ys.make_url('IWM',start_date='2010-2-1')
```

```python
# 'Adj Close' is Yahoo's adjustment to closing price for
# for splits and dividends. 'TotalValue' is another method.
IWM[['Close','TotalValue','Adj Close']].plot()
```
